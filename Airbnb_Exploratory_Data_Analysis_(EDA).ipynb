# Project Overview

```python
Business Context

Since 2008, Airbnb has enabled travelers and hosts to connect through a unique accommodation marketplace. With millions of listings worldwide, Airbnb generates large volumes of data that can be analyzed to support business decisions, understand customer and host behavior, optimize pricing, improve availability, and guide marketing strategies.

This project analyzes an Airbnb dataset containing ~49,000 listings and 16 features to extract meaningful insights using Exploratory Data Analysis (EDA).
```

```python
Problem Statement

The goal of this project is to explore and analyze Airbnb listing data to:

Understand pricing patterns

Analyze room types and neighborhood preferences

Study host behavior and listing availability

Identify trends and correlations among key variables
```

# Dataset Description

```python
The dataset contains the following key columns:

id – Unique listing ID

name – Name of the listing

host_id – Unique host ID

host_name – Name of the host

neighbourhood_group – Location group

neighbourhood – Area name

latitude – Latitude coordinate

room_type – Type of listing

price – Price per night

minimum_nights – Minimum nights required

number_of_reviews – Total number of reviews

last_review – Date of last review

reviews_per_month – Average monthly reviews

calculated_host_listings_count – Listings per host

availability_365 – Availability in a year
```

# Import Libraries

```python
# Importing required libraries for data analysis and visualization
import pandas as pd         # For data manipulation
import numpy as np          # For numerical operations
import matplotlib.pyplot as plt     # For basic visualizations
import seaborn as sns       # For advanced statistical plots


plt.style.use('default')
```

# Load Dataset

```python
df = pd.read_csv('Airbnb NYC 2019.csv')      # Loading the Airbnb NYC 2019 dataset
```

# Data Cleaning & Preprocessing

# Missing Value Analysis

```python
df.isnull().sum()
```

# Handling Missing Values

```python
# Handling missing values to ensure clean analysis
# Text columns are filled with 'Unknown'
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
df['last_review'] = df['last_review'].fillna('Unknown')

# Numerical column filled with 0 assuming no reviews
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

# Duplicate Records

```python
df.duplicated().sum()     # To check duplicate records
```

```python
df.drop_duplicates(inplace=True) # Removing duplicate records if any
```

# Exploratory Data Analysis (EDA)

# Visualization 1: Distribution of Room Types

```python
# Visualizing distribution of room types
plt.figure(figsize=(6,4))
sns.countplot(x='room_type', data=df)
plt.title('Distribution of Room Types')
plt.show()
```

# Visualization 2: Listings by Neighbourhood Group

```python
plt.figure(figsize=(6,4))
sns.countplot(x='neighbourhood_group', data=df)
plt.title('Listings by Neighbourhood Group')
plt.show()
```

# Visualization 3: Price Distribution

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['price'])
plt.title('Price Distribution')
plt.show()
```

# Visualization 4: Average Price by Room Type

```python
plt.figure(figsize=(6,4))
sns.barplot(x='room_type', y='price', data=df)
plt.title('Average Price by Room Type')
plt.show()
```

# Visualization 5: Average Price by Neighbourhood Group

```python
plt.figure(figsize=(7,4))
sns.barplot(x='neighbourhood_group', y='price', data=df)
plt.title('Average Price by Neighbourhood Group')
plt.show()
```

# Visualization 6: Availability Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df['availability_365'], bins=30)
plt.title('Availability Distribution')
plt.show()
```

# Visualization 7: Price vs Number of Reviews

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='price', y='number_of_reviews', data=df)
plt.title('Price vs Number of Reviews')
plt.show()
```

# Visualization 8: Correlation Heatmap (Numeric Features Only)

```python
# Selecting only numerical columns for correlation analysis
numeric_df = df.select_dtypes(include=['int64','float64'])

# Plotting correlation heatmap to identify relationships between variables
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
